## Greatest Blue Jays Teams of Forever

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data/core/Teams.csv")
```

```python
year = [1977, 2021]
redsox = False
yankees = False
```

```python
Tteams = df[df["teamID"] == "TOR"]
Bteams = df[df["teamID"] == "BOS"]
Yteams = df[df["teamID"] == "NYA"]
```

```python
plt.figure(figsize=(14, 8))
plt.title("Wins by Year")
plt.plot(Tteams["yearID"], Tteams["W"], marker="*")
if redsox == True:
    plt.plot(Bteams["yearID"], Bteams["W"], marker="x")
if yankees == True:
    plt.plot(Yteams["yearID"], Yteams["W"], marker=">")
plt.xlim(year)
plt.show()
```
